```python
import re

FILENAME = "scrach.md"

with open(FILENAME) as f:
    text = f.read()


# 日付の一覧を作成
rex = '#\s*(\d\d\d\d\/\d\d\/\d\d)\n+'
pattern = re.compile(rex)
date_list = re.findall(rex,text)

#セーブするファイル名の一覧を作成
file_name_list = []
for t in date_list:
    file_name_list.append(t.replace("/","-") + ".txt")

#本文を抽出
text_list = re.split(rex,text)
print(text_list[2])

#ファイルセーブ
for f_name in file_name_list:
    print(f_name)
    
    with open(f_name, mode='w') as f:
        f.write(text_list.pop(1)+"\n\n") # 日付をセーブ
        f.write(text_list.pop(1)) # 本文をセーブ
        
```
